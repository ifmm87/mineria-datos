```python
X = df1[nombres_varind]
y = df1['ViolentCrimesPerPop']
```

### Definición de sets de entrenamiento y test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# transformación normal
# Probamos una transformacion estandar para 
# normalizar la muestra
#scaler = StandardScaler()
#scaler.fit(X_train)
#X_train = scaler.fit_transform(X_train)
#X_test  = scaler.transform(X_test)
#X_train
#X_test

#scalery = StandardScaler()
#scalery.fit(y_train)
#y_train = scalery.fit_transform(y_train)
#y_trim
```

### Aplicación de la Técnica

```python
linreg = LinearRegression(normalize=True)
linreg.fit(X_train, y_train)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
zip(nombres_varind, linreg.coef_)
```

```python
# predicción
y_pred = linreg.predict(X_test)
# calculando residuals
residual = (y_test - y_pred)
```

```python
# histogramas
plt.hist(residual)
```

```python
stats.jarque_bera(residual)
```

```python
# calculate MAE using scikit-learn
print('Error absoluto medio:', mean_absolute_error(y_test, y_pred))
# calculate MSE using scikit-learn
print('Error cuadratico medio:', mean_squared_error(y_test, y_pred))
# coef de determinacion
print('coeficiente de determinación: ',r2_score(y_test, y_pred))
```
